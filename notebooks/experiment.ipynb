```python
df.columns = ['label', 'news_headline']
col = df.pop('label')
df['label'] = col

pd.set_option('display.max_colwidth',250)
df.sample(10)
```

```python
# Split the dataset
y = df['label']                 # target
X = df[['news_headline']]         # features

pd.set_option('display.max_colwidth',250)
X.sample(20)
```

```python
def clean_text(text):

    # make all the lowercaser
    text = text.lower()

    # remove puntuation & special characters
    text = re.sub()

    return text
```

```python
df['news_headline'] = df['news_headline'].apply(clean_text)
```
